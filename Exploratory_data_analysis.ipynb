## Question # 1: How many movies are released each year?

```python
# separate release year and title as two new columns
movies_df['year'] = movies_df.title.str.extract('(\(\d{4}\))',expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d{4})',expand=False)
movies_df['title_no_year'] = movies_df.title.str.replace('(\(\d{4}\))', '')
movies_df.head(10)
```

```python
movies_df.info()
```

```python
# print those movies with NaN in year
mask = movies_df['year'].isna()
movies_df[mask]
```

```python
# since there is not much missing year, we can manually add those missing releasing year
missing_years = ['1993',
                '2018',
                '2015',
                '1980',
                '2016',
                '2016',
                '2016',
                '2016',
                '1980',
                '2017',
                '2006',
                '2017',
                '2011',
                ]
movies_df.loc[mask, 'year'] = missing_years
```

```python
unique_years = np.array(sorted(movies_df['year'].astype(int).unique()))
print(unique_years)
```

```python
movies_df['year'] = movies_df['year'].astype(int)
```

```python
movies_df.describe()
```

```python
# get counts of movies numbers for each year
fig, ax = plt.subplots(figsize=(25,6))
movies_df.groupby(['year']).size().plot(kind='bar', ax=ax)
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Number of movies', fontsize=18)
ax.set_title('Movies released each year', fontsize=20)
plt.show()
```

## Questions # 2: How many movies of each gener released each year?

```python
# drop the original title column and change title_no_year to title
movies_df['title'] = movies_df['title_no_year']
movies_df.drop(['title_no_year'], axis=1, inplace=True)
movies_df.head()
```

```python
# get the list of unique genres
import itertools
genre_lists = movies_df.genres.str.split('|').tolist()
set(list(itertools.chain(*genre_lists)))
```

```python
# for convience, replace (no geners listed) with NA
movies_df['genres'] = movies_df.genres.str.replace('(no genres listed)', 'NA')
```

```python
# get the updated unique geners
genre_lists = movies_df.genres.str.split('|').tolist()
unique_genres = set(list(itertools.chain(*genre_lists)))
print(unique_genres)
```

```python
# add unique genres as additional columns 
movies_df = movies_df.join(movies_df.genres.str.get_dummies().astype(bool))
movies_df.head()
```

```python
# drop the original genres column
movies_df.drop(['genres'], inplace=True, axis=1)
```

```python
movies_df.head()
```

```python
# get a stacked bar plot for each genre over year
unique_genres = list(unique_genres)
fig, ax = plt.subplots(figsize=(25,12))
movies_df.groupby(['year'])[unique_genres].sum().plot(kind='bar', stacked=True, colormap='tab20', ax=ax)
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Number of movies', fontsize=18)
ax.set_title('Genres vs. year', fontsize=20)
```

```python
# or plot lines
fig, ax = plt.subplots(figsize=(25,12))
movies_df.groupby(['year'])[unique_genres].sum().plot(kind='line', colormap='tab20', ax=ax)
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Number of movies', fontsize=18)
ax.set_title('Genres vs. year', fontsize=20)
```

## Question # 5: How many rates for each year?

```python
# need to convert timestamp to year, month, date
ratings_df['year'] = pd.to_datetime(ratings_df['timestamp'], unit='s').dt.year
ratings_df['month'] = pd.to_datetime(ratings_df['timestamp'], unit='s').dt.month
ratings_df['day'] = pd.to_datetime(ratings_df['timestamp'], unit='s').dt.day
ratings_df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ratings_df.groupby(['year'])['rating'].size().plot(kind='bar', ax=ax)
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('# of ratings', fontsize=18)
ax.set_title('# of ratings for each year', fontsize=20)
plt.show()
```

## Questions # 6: What are the average ratings for each year?

```python
fig, ax = plt.subplots(figsize=(20,8))
ratings_df.groupby(['year'])['rating'].mean().plot(kind='bar', ax=ax)
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Average rating', fontsize=18)
ax.set_title('Average rating for each year', fontsize=20)
plt.show()
```

## Question # 10: What are the average ratings for each genre?

```python
# we need join two dataframes to obtain this insight
movies_df.head()
```

```python
ratings_df.head()
```

```python
ratings_movies = ratings_df.join(movies_df, how='left', on='movieId', lsuffix='_r', rsuffix='_m')
ratings_movies.head(10)
```

```python
genres_ratings = ratings_movies[unique_genres].multiply(ratings_movies['rating'], axis='index')
```

```python
genres_ratings.head()
```

```python
# when calculating the mean ratings of each genre, 0 values should be excluded
genres_ratings = genres_ratings.replace(0, np.NaN)
genres_ratings.head()
```

```python
# NaN is excluded by default
fig, ax = plt.subplots(figsize=(10,6))
genres_ratings.mean(axis=0).plot(kind='bar', ax=ax)
ax.set_ylabel('Ratings', fontsize=18)
ax.set_xlabel('Genres', fontsize=18)
ax.set_title('Average rating for different genres', fontsize=20)
```

## Question # 11: What are the number of ratings for each genres? 

```python
fig, ax = plt.subplots(figsize=(10,6))
genres_ratings.count().plot(kind='bar', ax=ax)
ax.set_ylabel('Counts', fontsize=18)
ax.set_xlabel('Genres', fontsize=18)
ax.set_title('# of ratings for different genres', fontsize=20)
```
